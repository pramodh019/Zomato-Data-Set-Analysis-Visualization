## Dropping duplicates

```python
df.duplicated().sum()
```

### cleaning rate column

```python
df['rate'].unique()
```

```python
df['rate'].nunique()
```

Removing "NEW" , "-" and "/5" from Rate Column

```python
def clean_rate(value):
    if value in ['NEW', '-']:
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate'] = df['rate'].apply(clean_rate)
df['rate'].head()
```

```python
df.isnull().sum()
```

### Filling Null Values in Rate Column with Mean

```python
from sklearn.impute import SimpleImputer
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy='mean')
impute.fit(df.iloc[:,3:4])  # Apply iloc to DataFrame
df.iloc[:,3:4] = impute.transform(df.iloc[:,3:4])  # Apply iloc to DataFrame
```

```python
df.info()
```

## Dropping NULL values

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

### Renaming columns for better reference

```python
df.rename(columns={'approx_cost(for two people)':'costfor2plates','listed_in(type)':'Type'},inplace= True)
df.head(3)
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Listed in(city) and location, both are there, lets keep only one.

```python
df= df.drop(['listed_in(city)'],axis=1)
```

```python
df['costfor2plates'].unique()
```

```python
def clean_comma(value):
    if ',' in value:
        value= value.replace(',','')
        return float(value)
    else:
        return float(value)


df['costfor2plates']=df['costfor2plates'].apply(clean_comma)

df['costfor2plates']
```

```python
df.head()
```

## Cleaning Rest Type Column

```python
rest_type = df['rest_type'].value_counts()
rest_type
```

```python
rest_types_lessthan1000 = rest_type[rest_type<1000]
rest_types_lessthan1000
```

### Making Rest Types less than 1000 in frequency as others

```python
def clean_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type']=df['rest_type'].apply(clean_rest_type)
df['rest_type'].value_counts()
```

## Cleaning Location Column

```python
location= df['location'].value_counts(ascending  = False)
```

```python
location
```

```python
location_lessthan300 = location[location<300]
location_lessthan300
```

```python
def clean_location(value):
    if (value in location_lessthan300): 
        return 'others'
    else:
        return value

df['location']= df['location'].apply(clean_location)
df['location'].value_counts()

        
```

### Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending= False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def clean_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines']=df['cuisines'].apply(clean_cuisines)
df['cuisines'].value_counts()
```
